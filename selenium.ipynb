```python
import random
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
```

```python

# Set up Chrome options to avoid detection
options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("start-maximized")
```

```python
# Use a realistic User-Agent string
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
options.add_argument(f"user-agent={user_agent}")

# Setup ChromeDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# Function to introduce randomized sleep
def random_sleep(min_seconds=2, max_seconds=5):
    time.sleep(random.uniform(min_seconds, max_seconds))

# Function to mimic human-like scrolling
def human_like_scroll(driver):
    scroll_pause_time = random.uniform(0.5, 1.5)
    screen_height = driver.execute_script("return window.innerHeight;")
    for i in range(1, 10):
        driver.execute_script(f"window.scrollTo(0, {screen_height * i});")
        time.sleep(scroll_pause_time)
    # Ensure the final scroll position is reached
    driver.execute_script(f"window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_pause_time)

# Function to click element using JavaScript
def click_element(driver, element):
    driver.execute_script("arguments[0].scrollIntoView(true);", element)
    driver.execute_script("arguments[0].click();", element)
```

```python

# Function to scrape ZoomInfo
def scrape_zoominfo():
    driver.get("https://www.linkedin.com/search/results/content/?keywords=ai4%202024&origin=GLOBAL_SEARCH_HEADER&sid=aOO")
    random_sleep(5, 10)  # Adjust sleep time as needed to ensure page loads

    while True:
        # Mimic human-like scrolling
        human_like_scroll(driver)

        # Extract company names and locations
        try:
            companies = WebDriverWait(driver, 10).until(
                EC.presence_of_all_elements_located((By.CSS_SELECTOR, ".company-name.link"))
            )
            locations = driver.find_elements(By.CSS_SELECTOR, ".row-cell.location")

            if not companies or not locations:
                raise Exception("Could not find companies or locations elements.")

            for company, location in zip(companies, locations):
                print(f"Company: {company.text}, Location: {location.text}")
                random_sleep(1, 3)  # Random sleep between extracting each company
        except Exception as e:
            print(f"Error extracting company names or locations: {e}")
            driver.save_screenshot('error_screenshot.png')  # Take a screenshot for debugging
            break

        # Navigate to next page
        try:
            next_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.CSS_SELECTOR, ".arrow-link.forward"))
            )
            click_element(driver, next_button)
            random_sleep(5, 10)  # Adjust sleep time as needed to ensure page loads
        except Exception as e:
            print(f"Error navigating to next page: {e}")
            driver.save_screenshot('next_page_error.png')  # Take a screenshot for debugging
            break
```

```python
# Start scraping
scrape_zoominfo()
```

```python
# BONUS: Unleash your wild side!
def wild_scrape(driver):
    while True:
        try:
            # Simulate random clicking on the page
            click_x = random.randint(1, 1920)
            click_y = random.randint(1, 1080)
            driver.execute_script(f"document.dispatchEvent(new MouseEvent('click', {{clientX: {click_x}, clientY: {click_y}}}));")
            print(f"Clicked at ({click_x}, {click_y})")
            random_sleep(1, 2)  # Add a small delay to avoid overwhelming the page
        except Exception as e:
            print(f"Error executing wild click: {e}")
            driver.save_screenshot('wild_click_error.png')  # Take a screenshot for debugging
            break
```
